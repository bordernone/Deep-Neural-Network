```python
import numpy as np
import tensorflow as tf
from sklearn import preprocessing
```

```python
dataset_inputs = np.loadtxt(fname="iris.csv", delimiter=",", usecols=(0,1,2,3))
dataset_targets = np.loadtxt(fname="iris.csv", delimiter=",", usecols=(5))

total_rows, total_cols = dataset_inputs.shape
```

```python
unique_cat_count = 3

one_hot_encoded_targets = tf.one_hot(dataset_targets, unique_cat_count, dtype=tf.int64)
```

```python
scaled_inputs = preprocessing.scale(dataset_inputs)
```

```python
dataset_tf = tf.data.Dataset.from_tensor_slices((scaled_inputs,one_hot_encoded_targets))
```

```python
num_train = 110
num_validation = 20
num_test = 20
```

```python
BUFFER_SIZE = 150
dataset_tf = dataset_tf.shuffle(BUFFER_SIZE)
```

```python
test_data = dataset_tf.take(num_test)

train_validation_data = dataset_tf.skip(num_test)

validation_data = train_validation_data.take(num_validation)

train_data = train_validation_data.skip(num_validation)
```

```python
batch_size = 10
test_data = test_data.batch(num_test)
validation_data = validation_data.batch(num_validation)
train_data = train_data.batch(batch_size)
```

```python
validation_inputs, validation_targets = next(iter(validation_data))
```

```python
input_size = total_cols
output_size = unique_cat_count
hidden_layer_size = 100

model = tf.keras.Sequential([
    tf.keras.layers.InputLayer(input_shape=(total_cols,)),
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(output_size, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
NUM_EPOCHS = 100

early_stopper = tf.keras.callbacks.EarlyStopping()

model.fit(train_data, epochs=NUM_EPOCHS, validation_data=(validation_inputs, validation_targets), verbose=2, callbacks=[early_stopper])
```

```python
test_loss, test_accuracy = model.evaluate(test_data)
```
